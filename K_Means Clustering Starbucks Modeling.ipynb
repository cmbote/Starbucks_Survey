K-Means Clustering

```python
encoder = LabelEncoder()
```

```python
for column in object_columns:
    data[column] = encoder.fit_transform(data[column]) 
```

```python
data.head()
```

Picking a randon number for K (between 3 and 5)

```python
X_train =  data.values
```

```python
kmeans = KMeans(n_clusters=4, random_state = 42)
kmeans = kmeans.fit(X_train)
predictions = kmeans.predict(X_train)

clusters = kmeans.labels_
inertia = kmeans.inertia_
print("The clusters are: ", clusters)
print("The Inertia is : ", inertia)
```

```python
no_of_clusters = range(2,21)
inertia=[]

for f in no_of_clusters:
  kmeans = KMeans(n_clusters=f, random_state=2)
  kmeans = kmeans.fit(X_train)
  u = kmeans.inertia_
  inertia.append(u)
  print('The inertia for :',f,'Cluster is:',u)
```

Elbow Plot

```python
fig, (ax1) = plt.subplots(1, figsize=(16,6))
xx = np.arange(len(no_of_clusters))
ax1.plot(xx, inertia)
ax1.set_xticks(xx)
ax1.set_xticklabels(no_of_clusters, rotation = 'vertical')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia Score')
plt.title('Inertia Plot per k')

plt.annotate('Selecting the number of clusters', xy=(2, inertia[2]), 
             xytext=(2, inertia[10]), 
            arrowprops=dict(facecolor='red', shrink=0.05))
```

```python
unique, counts = np.unique(predictions, return_counts=True)
counts = counts.reshape(1,4) #1 row, 4 new columns

clf = pd.DataFrame(counts, columns = ['Cluster 0','Cluster 1','Cluster 2','Cluster 3'])

clf
```
